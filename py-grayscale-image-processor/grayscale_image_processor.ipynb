# **Conversão de imagem RGB em imagem Grayscale (níveis de cinza)**

**Instalação das Bibliotecas** 

É realizada a instalação das seguintes bibliotecas:
1.	 OpenCV, para processamento de imagens.
2.	 NumPy, para trabalhar com grandes computações numéricas

Também é importada da biblioteca do Google Colab Patches o método `cv2_imshow` da biblioteca OpenCV, responsável por exibir a imagem, já que o da biblioteca original ocasiona erros no Google Colab.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

**Leitura e exibição da imagem** 

É criada uma variável chamada `img`. Através do método `imread` da biblioteca OpenCV, essa variável recebe o valor da imagem a ser manipulada, passando o diretório dela. Após isso, é utilizado o método imshow nessa variável para exibi-la.

```python
img = cv2.imread('./assets/img-01.jpg',1)
cv2_imshow(img)
```

img-01.jpg

**Separação das matrizes dos valores RGB**

São criadas três variáveis, chamadas de `b` (blue), `g` (green) e `r` (red). Através do método `split` da biblioteca OpenCV, cada variável recebe o valor de uma matriz respectiva a cada cor da coloração RGB. Ou seja, a variável `b` recebe as matrizes responsáveis pela matriz de coloração azul (blue) de uma imagem, e assim por diante.

```python
B, G, R = cv2.split(img)
```

**Conversão da imagem e resultado final**

São constatados dois tipos de conversões:

1. **Aplicação da conversão básica:**



> Primeiramente, é criada uma variável `img_grayscale_basic`. Essa variável recebe o resultado do seguinte cálculo: o valor das matrizes responsáveis por cada coloração RGB são somadas e no final, o valor resultante é dividido por 3. Após isso, pelo método `array` da biblioteca NumPy, essa mesma variável recebe um array das matrizes resultantes do cálculo anterior e é passado um tipo de valor inteiro de 8 bits (`uint8`), atribuindo números inteiros de 0 a 255.

Após isso,  basta utilizar o método `imshow` da biblioteca OpenCV para exibir a variável responsável pela imagem já convertida.



```python
img_grayscale_basic = (B+G+R)/3

img_grayscale_basic = np.array(img_grayscale_basic, dtype=np.uint8)

cv2_imshow(img_grayscale_basic)
```

img-02.png

2. **Aplicação da conversão ponderada**

> Cada cor RGB possui um percentual de luminância. Essa conversão leva em conta esse percentual, por isso é chamada de ponderada. Primeiramente, é criada uma variável `image_grayscale_pondered`. A partir do método `array` da biblioteca NumPy, a variável recebe um array de novas matrizes a partir do cálculo realizado com as proporções de luminância RGB e passando um tipo de valor inteiro de 8 bits (`uint8`), atribuindo números inteiros de 0 a 255.

Após isso,  basta utilizar o método `imshow` da biblioteca OpenCV para exibir a variável responsável pela imagem já convertida.



```python
img_grayscale_pondered = 0.299*B+0.587*G+0.114*R

img_grayscale_pondered = np.array(img_grayscale_pondered, dtype=np.uint8)

cv2_imshow(img_grayscale_pondered)
```
